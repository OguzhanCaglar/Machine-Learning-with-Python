__Overfitting:__ Kabaca modelin "eğitim veri kümesinde iyi performans gösterdiği halde sınama ve test veri kümelerinde çok düşük bir performans göstermesi" anlamına gelmektedir. Sinyal-gürültü (signal-noise). Eğitim verilerimizde öğrenilmesi gereken kalıplar "sinyallerdir" ancak bu sinyallerin arasında "gürültüler" de vardır. Eeğer modelimiz sinyalleri değil de gürültüleri öğrenme eğilimindeyse overfitting oluşmaktadır.

Nedenleri: 
- Eğitim verilerinin uygun biçimde oluşturulamaması. Eğitim kümesi yanlı (biased) şekilde seçilmiş olabilir. Varyans'ın düşük olması. 
- modelim doğru biçimde oluşturulamaması 
- Fazla epoch uygulamak. Uygun bir epoch'ta eğitim kesilmeli (Early stopping)
- Modelin çok fazla parametreye sahip olması.

Nasıl Anlarız?:
- Eğitim verilerinin verdiği metrik değerler iyileşirken bu iyileşme sınama verilerinde gözükmüyorsa yani eğitim verilerinni kendisinden elde edilen başarı sınama sırasında ortaya çıkmıyorsa bu durumda bir "overfitting" durumu söz konusu olabilir. 

__Underfitting:__ Overfitting olgusunun tersine "underfitting" denilmektedir. Underfitting de yine modelin yanlış oluştrulmasından ve eğitim veri kümesindeki sorunlardan kaynaklanıyor olabilir. Örneğin özelliklerin yanlış ve eksik seçilmesi, modeldeki nöronların sayısının azlığı, yanlış optimizasyon algoritmalarının ve loss fonksiyonlarının kullanılması modelin yanlış oluşturulduğuna ilişkin akla gelen en önemli unsurlardır. Düşük varyans overfitting durumuna yol açarken yüksek varyanstan da underfitting durumuna yol açabilmektedir.

### Regresyon Kavramı ve Regresyon Modellerinin

Girdi ile çıktı arasında ilişki kurma sürecine istatistikte "regresyon" denilmektedir. Girdiyle çıktı arasındaki ilişki bir fonksiyonla ifade edilebildiğine göre regresyon sürecini de biz aslında girdi ile çıktı arasındaki ilişkiyi belirten uygun bir fonksiyonun bulunması süreci olarak ele alabiliriz. y = f(x). biçiminde Regresyon bir f fonksiyonun bulunması sürecidir. 

Girdi ile çıktı arasında ilişki kurmanın amacı kestirimde bulunmaktatır. Girdiler yoluyla çıktıların tahmin edilmesine yönelik problemlere "regresyon problemleri" diyeceğiz.

Regresyon modelleri ve problemleri istatistikte çeşitli biçimlerde sınıflandıırılabilmektedir. 

__Doğrusal Regresyon (Linear Regression)__

Girdi çıktı arasındaki ilişkinin doğrusal olduğu kabulüyle yapılan regresyon sürecine denilmektedir. Yani doğrusal regresyonla girdi ile çıktı arasında doğrusal bir fonksiyon bulunmaya çalışılır.



Bu doğru aslında noktaların doğruya uzaklıklarının en küçük olduğu doğrrudur. En küçük kareler yöntemi denilmektedir. Doğrusal regresyonda girdi (yani bağımsız değişken) bir tane ve çıktı da (bağımlı değişken) bir tane ise buna "basit doğrusal regresyon" denilmektedir. Eğer doğrusal regresyonda girdiler (yani bağımsız değişkenler) birden fazla ise buna "çoklu doğrusal regresyon (multiple linear regression) denilmektedir. 

f(x) = a1*x1 + a2*x2 + a3*x3 + .... + an*xn + b

y = ax + b

__Kodu__

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)

y_pred = model.predict(x_new)

__Polinomsal Regresyon (Polynomial Regression):__

Bağımsız değişkenlerin herhangi birinin üssü 1'den büyükse bu tür regresyon modellerine polinomsal regresyon modelleri denilmektedir.





__Kodu__

from sklearn.linear_model import PolynomialRegression

model = PolynomialRegression(degree=2)

model.fit(x,y)

y_pred = model.predict(x_new)

Doğrusal olmayan regresyon denildiğinde katsayılarla değişkenler değiştirildiğinde yine doğrusal olmayan fonksiyonların elde edildiği regresyonlar anlaşılmaktadır. üstel ifadeler, logairtmik ifadeler, 



__Karar Ağacı Regresyonu:__

Karar ağacı, her bir iç düğümün bir öznitelik üzerinde bir testi belirttiği, her bir dalın testin bir sonucunu temsil ettiği ve her bir yaprak düğümünün (terminal) bir sınıf etiketi tuttuuğu akış şeması benzeri bir ağaç yapısıdır. Sürekli bir sonucu tahmin etmek için bir karar ağacını modellemek mümkündür.



from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(x, y)

y_pred = model.predict(x,y)

__Random Forest Regresion__

Birden fazla karar ağacı ile "boostrap ve aggregation" adı verilen bir teknik kullanılarak hem regresyon hem de sınıflandırma görevlerini gerçekleştirebilen bir topluluk tekniğidir. Temel fikir, nihai çıktıyı belirlemede tek tek karar ağaçlarına güvenmek yerine birden fazla karar ağacını birleştirmektir.



from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100)

model.fit(x, y)

y_pred = model.predict(x,y)

__Destek Vektör Regresyonu (SVR)__

Regresyon görevleri için kullanılan bir tür destek vektör makinesidir (svm). Belirli bir girdi değeri için sürekli çıktı eğerini en iyi şekilde tahmin eden bir fonksiyon bulmaaya çalışıyor. SVR hem doğrusal hem de doğrusal olmayan "kernel" adında fonksiyonlar kullanır. Doğrusal bir kernel (çekirdek) iki giriş vektörü arasındaki basit bir nokta fonksiyonudur. Doğrusal olmayan bir çekirdek ise verilerdeki karmaşık desenleri yakalyabilen daha karmaşık bir fonksğiyondur. Kernel seçimi, verilerin özelliklerine ve görevin karmaşıklığına bağlıdır. 



from sklearn.svm import SVR

model = SVR(kernel = 'linear') # kernel = 'poly', kernel = 'rbf'

model.fit(x,y)

y_pred = model.predict(x_new)

### Yapay Sinir Ağları ile Oluşturulan Regresyon Modellerinde Model Parametrelerinin Belirlenmesi

- Sınıflandırma problemlerinde genellikle saklı katmanlar için "relu" aktivasyon fonksiynu tercih edilir. Çıktı katmanı için ikili sınıflandırma problemlerinde "sigmoid", çoklu sınıflanımra problemlerinde "softmax", lojistik olmayan regresyon problemlernde ise "linear" aktivasyon fonksiyonları tercih edilmektedir.

- Modeldeki optimizer algoritması için genellikle "rmsprop", "adam", ve "sgd" seçilmektedir. 

__ÖDEV:__ Bu optimizasyon algoritmalarını araştırın.

- Modelin loss fonksiyonu için genellikle ikili sınıflandrma modelleri için "binary_crossentropy", çoklu sınıflandırma modelleri için "categorical_crossentropy" ve lojistik olmayan regresyon modelleri için de "mse" tercih edilir.

- Sınama işlemi için kullanılacak metrikler de genellikle genellikle ikili sınıflandrma modelleri için "binary_accuracy", çoklu sınıflandırma modelleri için "categorical_accuracy", lojistik olmayan regesyon modelleri için ise "mae" seçilir.

- Model katmanları "dense" (yani önceki katmandaki tüm nöronların çıktılarının sonraki katmandaki tüm nöronlara bağlanması biçiminde) biçimde oluşturulur. 

- Katman nöron sayıları, 32, 64, 100, 200 gibi orta değerlerde tutulmaktadır.

- Önce tek saklı katman denenir, eğer performans düşük kalırsa katman sayısı ikiye, üçe, dörde duruma göre daha yükske derecelere çıkartılır. Pek çok problem için iki saklı katman yeterli olmaktadır. 

- Fazla miktarda katmanla hala iyileşme sağlanamıyorsa katmanlardaki nöron sayıları arttırılır.
