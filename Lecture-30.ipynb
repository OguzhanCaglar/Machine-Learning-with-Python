### Uygulama:

10 adet zellik ve bir adet hedef deikeni olan diabetes veri setini kullanalm. Bu uygulamada diayabete etkisi zellikerlden biri seilerek bir regresyon modeli oluturulmaktadr. Olutuurlan regresyon modeli verilen deere gre bir diyabet deeri tahmini yapmaktadr

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

pd.DataFrame(data = diabetes_X).info()
diabetes_X = diabetes_X[:, np.newaxis, 2] # yalnz bir zellik seiliyor.
```

```python
diabetes_X
```

```python
#veri setini eitim ve test olarak blelim.
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
# hedef deiken eitim ve test olarak blnr.
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
# zellik stunu ve diyabet deeri arasndaki ilikiyi grebilmek ini veriyi grselletirelim.

plt.scatter(diabetes_X, diabetes_y)
plt.title("Diyabet Grafik")
plt.xlabel("zellik")
plt.ylabel("Diyabet Deeri")
plt.grid(True)
plt.show()
```

```python
# modeli eitelim
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# eitilen m odelin test veri setindeki zelliklerini kullnaarak tahmin yapalm
diabetes_y_pred = regr.predict(diabetes_X_test)
```

```python
diabetes_y_pred
```

```python
print('Sabit (bias): \n', regr.intercept_)
print('Eim: \n', regr.coef_)
```

```python
# Gerek deerler:
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
# Regresyon dorusu:
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue')
```

```python
# performas lm
print("Ortalama Kare Hata: {}".format(mean_squared_error(diabetes_y_test, diabetes_y_pred)))
print("R-kare: {}".format(r2_score(diabetes_y_test, diabetes_y_pred)))
```
