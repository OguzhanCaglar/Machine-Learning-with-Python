## 1) Denetimli (Gözetimli Öğrenme Algoritmaları)

Denetimli öğrenme modelleri sınıflandırma ve regresyon problemlerinin çözümünde kullanılır. Sınıflandırma bir örneğin ait olduğu sınıfı, kategoriyi belirlemek için kullanılduığını söylemiştik. İkili sınıflandırma, çok sıınflı sınıflandırma (multi class classification). Sınıf adlarının her birine etiket deniliyordu. Sınıflandırma için kullaınlacak veri setlerinde sınıf etiketlerinni bulunduğu bir sütun olmalıdır.
Regresyonda ise bu sütun sürekli sayısal bir değer olmalıdır.

İris veri seti çiçeklerle ilgili bir veri seti. İris veri setindeki tür sütunu çiçeğin türüne (sınıfına) ait katıegiorik bir veri içerir. Setosa, versicolor, viriginica çiçeğin türünü (sınıfını) belirten etiketlerdir. Veri setinde tür sütununda üç kategori bulunduğu için çok sınıflı sınıflandırmaya örnek olarak verilebilir. Çiçeğe ait özellikler (sütunlar) çanak yaprak uzunluğu, çanak yaprak genişliği , taç yaprak uzunluğu ve taç yaprak genişliğidir. Bu özelliklere bağımsız değişken adı verilir. Çiçeğin türü ise bağımlı değişkendir.

### Denetimli (Gözetimli) Öğrenmede Model Seçimi?

# 3) Regresyon Analizi ve Türleri

Regresyon analizi özellik (bağımsız değişken) veya özelliklerle bir bağımlı değişken arasındaki ilişkiyi belirlemek için kullanılır. Regresyon için kıdem-ücret, çalışma süresi-alınan not ve kan değerleri-hastalık durumu gibi örnekler verilebilir. Regresyon analizi gözetimli öğrenme yöntemlerindendir. Doğrusal ve doğrusal olmayan regresyon modelleri bulunmaktadır.

## 3.1) Doğrusal ve Doğrusal Olmayan Regresyon

Doğrusal regresyon düz bir çizgiyle ifade edilebilen bir model türüdür. basit doğrusal regresyon ve çoklu doğrusal regresyon olmak üzere iki tür doğrusal regresyon vardır. Doğrusal olmayan regresyon düz bir çizgi ile ifade edilemeyen modeller için kullanılır. Polinomsal regresyon ve lojistik regresyon, doğrusal olmayan regresyon türleridir.

## 3.2) Basit Doğrusal Regresyon

İki değişken arasındaki ilişkiyi tanımlamak için kullanılan doğrusal bir modelin tanımıdır. Basit doğrusal regresyonda değişkenler de bağımlı değişşken ve bağımsız değişkendir. Doğrusal regresyonnu amacı gerçek değerlere en yakın doğruyu çizmektir. Doğrunun gerçek değerlere uzaklıkları hesaplanır. Ortalama uzaklık değerleri (hata) en düşük doğrunun formülü bulunur. 

y = Bo + B1 . X + e
Bo: Doğrunun y eksinini kestiği nokta ve regresyon sabitidir.
B1: Doğrunun eğimi ve katsayı
X: Bağımsız değişkenin değeri
e: Hata değeri



## 3.3) Çoklu Doğrusal Regresyon

Birden fazla bağımsız değişkenin ve bir bağımlı değişkenni bulunduğu regresyon modeli çoklu doğrusal regresyon modelidir. Çoklu doğrusal regresyon formülü:

y = y = Bo + B1 . X1 + B2 . X2 + ... + Bn . Xn + e

## 3.4) Polinomsal (Polinominal) Regresyon (Polynomial Regression)

Değişken arasındaki ilişkinin doğrusal olmadığı durumlarda Polinomsal regresyon analizi yapılmaktadır. Bu modelde denklemdeki parametre sayısı ile veri kümesindeki değişken sayısı doğrudan ilişkili olmayabilir. Doğrusal regresyon farklı olarak bir doğru yerine bir polinom grafiği çizilmektedir. Polinom derecesi oluşturulan modelin uyumunu etkiler.



Buradaki h polinom derecesini göstermektedir. İkinci derece polinoma kuadratik, üçşüncü derece polinoma kübik ve dördünce polinoma kuartik denir.

## 3.5) Lojistik (Logistic) Regresyon

Sınıflandırma için kullanılan denetimli bir makine öğrenmesi algoritmasıdır. Bir veya daha fazla bağımsız değişken ile bağımlı dfeğişken arasındaki ilişkiyi modellemek için kullanılır. Lojistik regresyon ikili sınıflandırma için bir regresyon fonksiyonu oluşturur. Bu fonksiyon ikili bir sınıf için 0 ve 1 çıktıları üretir. Lojistik reggresyona cinsiyet tahmini (erkek, kadın) ve hastalık durumu (pozitif, negatif) örnek verilebilir. 



## 3.6) Tahmin Modellerinin Değerlendirilmesi En Küçük Kareler Yöntemi

Regresyon yöntemlerinin başarısı ortalama kare hata (MSE) ölçütü ile karşılaştırılır. En küçük kareler yöntemi ortalama kare hatanın (MSE) minimize edilmesine dayanmaktadır. En küçük karaler yönteminfde hatayı minimize edecek regresyon çizgisini elde etmek amaçlanmaktadır:



#### Korelasyon ve regresyon ilişkisi

Regresyon değişkenler arasındaki neden-sonuç ilişksinii araştırırken korelasyon iki değişken arasında ilişkinin yönünü ve şiddetini ölçer. Yani regresyonda iki değişken biri dğiğerine bağımlıyken korelasyonda iki değişkenin valrığı birbirinden bağımsızdır. Örneğin yemek yeme ile kilo alma arasındaki ilişki regresyona örnek verilebilecekken günlük uyku süresi ile TV izleme süresi arasındaki ilişki koerlasyona örnek verilebilir.

R2 (Square): Doğrusal modellerde en çok kullanılan metriklerden biridir. Bir bağımsız değişkenle bir bağımlı değişkenin tahmin edildiği modellerde kullanılır. Genel olarak R2 en yüksek 1 değerini alabilir. Genel olarak R2 değeri ne kadar büyük olursa model o kadar başarılıdır. 

Düzeltilmiş R2: Modelde bağımsız değişken ssayısı birden fazlaysa R2 değeri yalnıtıxc olabilir. Bu tür modellerde düzeltilmiş R2 değeri kullanılır. 



### Uygulama:

10 adet özellik ve bir adet hedef değişkeni olan diabetes veri setini kullanalım. Bu uygulamada diayabete etkisi özellikerlden biri seçilerek bir regresyon modeli oluşturulmaktadır. Oluştuurlan regresyon modeli verilen değere göre bir diyabet değeri tahmini yapmaktadır

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

pd.DataFrame(data = diabetes_X).info()
diabetes_X = diabetes_X[:, np.newaxis, 2] # yalnız bir özellik seçiliyor.
```

```python
diabetes_X
```

```python
#veri setini eğitim ve test olarak bölelim.
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
# hedef değişken eğitim ve test olarak bölünür.
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
# özellik sütunu ve diyabet değeri arasındaki ilişkiyi görebilmek içni veriyi görselleştirelim.

plt.scatter(diabetes_X, diabetes_y)
plt.title("Diyabet Grafik")
plt.xlabel("Özellik")
plt.ylabel("Diyabet Değeri")
plt.grid(True)
plt.show()
```

```python
# modeli eğitelim
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# eğitilen m odelin test veri setindeki özelliklerini kullnaarak tahmin yapalım
diabetes_y_pred = regr.predict(diabetes_X_test)
```

```python
diabetes_y_pred
```

```python
print('Sabit (bias): \n', regr.intercept_)
print('Eğim: \n', regr.coef_)
```

diyabet_değeri = 152.92 + (938.24) * özellik

```python
# Gerçek değerler:
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
# Regresyon doğrusu:
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue')
```

```python
# performas ölçümü
print("Ortalama Kare Hata: {}".format(mean_squared_error(diabetes_y_test, diabetes_y_pred)))
print("R-kare: {}".format(r2_score(diabetes_y_test, diabetes_y_pred)))
```
