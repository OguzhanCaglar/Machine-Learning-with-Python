```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# replace categorical values with numerical values
dataset.replace({'sex' : {'female' : 1, 'male' : 0}}, inplace=True)
dataset.replace({'smoker' : {'yes' : 0, 'no' : 1}}, inplace=True)
dataset.replace({'region' : {'southeast' : 0, 'southwest' : 1, 'northwest' : 2, 'northeast' : 3}}, inplace=True)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
dataset['sex'].value_counts().plot(kind='pie', colors=['gold', 'pink'], shadow = True, autopct='%.2f')
plt.title("Distribution Of Sex", fontsize=15)

plt.subplot(1, 2, 2)
sns.countplot(x = dataset['sex'], palette = 'pink')
plt.title('Distribution of Sex (Counplot)', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
dataset['smoker'].value_counts().plot(kind='pie', colors=['lightgreen', 'violet'], shadow = True, autopct='%.2f')
plt.title("Distribution Of Smoker", fontsize=15)

plt.subplot(1, 2, 2)
sns.countplot(x = dataset['smoker'], palette = 'crest')
plt.title('Distribution of Smoker (Counplot)', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
dataset['region'].value_counts().plot(kind='pie', colors=['lightgreen', 'violet', 'yellow', 'gold'], shadow = True, autopct='%.2f')
plt.title("Distribution Of Region", fontsize=15)

plt.subplot(1, 2, 2)
sns.countplot(x = dataset['region'], palette = 'Purples')
plt.title('Distribution of Region (Counplot)', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(13, 6))
sns.countplot(x = dataset['children'], palette= 'YlOrBr')
plt.title('Distribution of Children (Counplot)', fontsize = 15)
plt.xlabel('Children', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(13, 6))
sns.heatmap(dataset.corr(), annot=True, cmap='Reds', linewidths=2)
plt.title("Correlation Metrics Of Numerical Features", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(14, 10))

for i, cols in enumerate(['age', 'bmi', 'children', 'charges']):
    plt.subplot(2, 2, i+1)
    sns.histplot(x = cols, data = dataset, kde = True, color = 'skyblue')
    plt.title(f'Distribution of {cols}', fontsize = 14)
    plt.tight_layout(pad=4.0)

plt.show()
```

```python
X = dataset.drop('charges', axis = 1)
Y = dataset['charges']
```

```python
X
```

```python
Y
```

```python
#(%80 train, %20 test)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn import metrics

regressor = LinearRegression()
regressor.fit(X_train, Y_train)

#prediction
training_data_prediction = regressor.predict(X_train)
```

```python
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value: ', r2_train)
```

```python
test_data_prediction = regressor.predict(X_test)

r2_test = metrics.r2_score(Y_test, test_data_prediction)
```

```python
print('R Squared value: ', r2_test)
```
